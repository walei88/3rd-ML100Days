# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = '../data/home-credit-default-risk/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
# 先看一下數據中有哪寫類型
at_df = pd.DataFrame(app_train) #製作app_train的DataFrame
app_train_dtype = at_df.dtypes.reset_index(0)
app_train_dtype.columns = ['count','dType']
app_train_dtype = app_train_dtype.groupby('dType').aggregate('count').reset_index(0)
app_train_dtype
```

```python
# 先篩選數值型的欄位
"""
YOUR CODE HERE, fill correct data types (for example str, float, int, ...)
"""
dtype_select = [np.dtype('float64')]
numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(dtype_select))])

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
#apply(lambda)的用法：https://www.itread01.com/content/1548122409.html
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2 ))])
print("Numbers of remain columns : %d" % len(numeric_columns))
```

```python
numeric_columns = [numeric_columns[0],numeric_columns[4],numeric_columns[56]]
```

```python
# 檢視這些欄位的數值範圍
for col in numeric_columns:
    plt.hist(app_train[col],bins=20)
    plt.title(col)
    plt.show()
```

```python
# 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

# AMT_INCOME_TOTAL：直方圖右側的空白很多，代表可能有離群值。

# REGION_POPULATION_RELATIVE：
#右側可以看到在X軸的0.07與0.04~0.05的地方，有2組數據與其他組數據之間沒有連續。
#此數據與客戶所居住的人口密集度有關，所以應確認一下不連續的地方有無異常之處。

# OBS_60_CNT_SOCIAL_CIRCLE：直方圖右側的空白很多，代表可能有離群值。
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

其他參考資料：https://medium.com/ai反斗城/exploratory-data-analysis-探索資料-ecdf-7fa350c32897

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
cdf = pd.DataFrame(sorted(app_train['AMT_INCOME_TOTAL']))
cdf.index = sorted(app_train['AMT_INCOME_TOTAL'])
cdf[0] = np.arange(1,len(cdf)+1)

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```

```python
# 最大值落在分布之外
print(app_train['REGION_POPULATION_RELATIVE'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
cdf = pd.DataFrame(sorted(app_train['REGION_POPULATION_RELATIVE']))
cdf.index = sorted(app_train['REGION_POPULATION_RELATIVE'])
cdf[0] = np.arange(1,len(cdf)+1)


plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['REGION_POPULATION_RELATIVE'].hist()
plt.show()

app_train['REGION_POPULATION_RELATIVE'].value_counts()

# 就以這個欄位來說，雖然有資料掉在分布以外，也不算異常，僅代表這間公司在稍微熱鬧的地區有的據點較少，
# 導致 region population relative 在少的部分較為密集，但在大的部分較為疏漏
```

```python
# 最大值落在分布之外
print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
cdf = pd.DataFrame(sorted(app_train['OBS_60_CNT_SOCIAL_CIRCLE']))
cdf = cdf.dropna().sort_values(by=0)
cdf.index = cdf[0]
cdf[0] = np.arange(1,len(cdf)+1)

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min() * 0.95, cdf.index.max() * 1.05])
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['OBS_60_CNT_SOCIAL_CIRCLE'].hist()
plt.show()
print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].value_counts().sort_index(ascending = False))
```
