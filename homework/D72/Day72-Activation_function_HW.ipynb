# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''

#定義ReLU的 數學函數表示方式
def ReLU(x):
    y = []
    for i in x:
        if i < 0:
            y.append(0)
        else:
            y.append(i)
    return y

#定義ReLU的 一階導數表示方式
def dReLU(x):
    y = []
    for i in x:
        if i < 0:
            y.append(0)
        else:
            y.append(1)
    return y
```

```python
# 給定x的範圍為-10 ～ 10，等距給出100個sample
x = plt.linspace(-10, 10, 100)

# 繪製ReLU函數
plt.plot(x, ReLU(x), 'b')
plt.grid()
plt.title('ReLU Function')
plt.text(3, 1.5, r'$\sigma(x)= x(x>0)$', fontsize=15)
plt.text(3, 0, r'$\sigma(x)=0 (x<0)$', fontsize=15)
plt.xlim(-10,10)

plt.show()
```

```python
# 繪製ReLU一階導數
plt.plot(x, dReLU(x), 'r')
plt.grid()
plt.title('dReLU Function')
plt.text(3, 0.55, r'$\sigma\'(x)= 1(x>0)$', fontsize=15)
plt.text(3, 0.40, r'$\sigma\'(x)=0 (x<0)$', fontsize=15)
plt.xlim(-10,10)

plt.show()
```
