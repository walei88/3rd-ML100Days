### 用 digits (dataset.load_digits()) 資料嘗試跑 t-SNE

```python
from sklearn import datasets

digits = datasets.load_digits()
X = digits.data
y = digits.target
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import offsetbox
from sklearn import manifold

%matplotlib inline
```

```python
# 筆記：n_components是降維之後的維度，本例中原本維度是64維，降維之後的維度是2維。
# 筆記：learning_rate是學習率，表示梯度下降的快慢，默認為200，建議取值在10到1000之間。
# 筆記：early_exaggeration表示嵌入空間簇間距的大小，默認為12，該值越大，可視化後的簇間距越大。
# 參考資料：https://blog.csdn.net/hustqb/article/details/80628721
tsne = manifold.TSNE(n_components=2, random_state=0, init='pca', learning_rate=300., early_exaggeration=20.)
```

```python
# Scale and visualize the embedding vectors
def plot_embedding(X, title=None):
    x_min, x_max = np.min(X, 0), np.max(X, 0) # np.min(X, 0) 其中0是代表axis=0，也就是根據行來取min
    X = (X - x_min) / (x_max - x_min)

    plt.figure()
    ax = plt.subplot(111)
    for i in range(X.shape[0]):
        plt.text(X[i, 0], X[i, 1], str(y[i]),
                 color=plt.cm.Set1(y[i] / 10.),
                 fontdict={'weight': 'bold', 'size': 9})

    # 筆記：hasattr(object, name)，如果對象有該屬性返回 True，否則返回 False。
    # 參考資料：https://www.runoob.com/python/python-func-hasattr.html
    if hasattr(offsetbox, 'AnnotationBbox'): 
        # only print thumbnails with matplotlib > 1.0
        shown_images = np.array([[1., 1.]])  # just something big
        for i in range(X.shape[0]):
            dist = np.sum((X[i] - shown_images) ** 2, 1)
            if np.min(dist) < 4e-3:
                # don't show points that are too close
                continue
            shown_images = np.r_[shown_images, [X[i]]]
            imagebox = offsetbox.AnnotationBbox(
                offsetbox.OffsetImage(digits.images[i], cmap=plt.cm.gray_r),
                X[i])
            ax.add_artist(imagebox)
    plt.xticks([]), plt.yticks([])
    if title is not None:
        plt.title(title)

# t-SNE embedding of the digits dataset
print("Computing t-SNE embedding")

X_tsne = tsne.fit_transform(X)

plot_embedding(X_tsne,
               "t-SNE embedding of the digits")

plt.show()
```

```python
tsne = manifold.TSNE(n_components=2, random_state=0, init='pca', learning_rate=300., early_exaggeration=120.)

# t-SNE embedding of the digits dataset
print("Computing t-SNE embedding")

X_tsne = tsne.fit_transform(X)
plot_embedding(X_tsne,
               "t-SNE embedding of the digits")
plt.show()
```

early_exaggeration 不能設置太大，否則很容易會分錯的情形更為嚴重。

```python
tsne = manifold.TSNE(n_components=2, random_state=0, init='pca', learning_rate=200., early_exaggeration=12.)

# t-SNE embedding of the digits dataset
print("Computing t-SNE embedding")

X_tsne = tsne.fit_transform(X)
plot_embedding(X_tsne,
               "t-SNE embedding of the digits")
plt.show()
```
