```python
# 載入基礎套件
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

### 弱相關

```python
# random.seed()的應用
# https://blog.csdn.net/linzch3/article/details/58220569
np.random.seed(1)

# 隨機生成兩組 1000 個介於 0~50 的數的整數 x, y, 看看相關矩陣如何
x1 = np.random.randint(0,50,1000)
y1 = np.random.randint(0,50,1000)

# 呼叫 numpy 裡的相關矩陣函數 (corrcoef)
np.corrcoef(x1,y1)
```

```python
# 將分布畫出來看看吧
plt.scatter(x1,y1)
```

### 正相關

```python
# np.random的用法
# https://docs.scipy.org/doc/numpy-1.16.0/reference/routines.random.html
# np.random.normal的用法
# https://docs.scipy.org/doc/numpy-1.16.0/reference/generated/numpy.random.normal.html#numpy.random.normal
# https://www.jianshu.com/p/214798dd8f93

np.random.seed(1)

# 隨機生成 1000 個介於 0~50 的數 x
x2 = np.random.randint(0,50,1000)

# 這次讓 y 與 x 正相關，再增加一些雜訊
y2 = x2 + np.random.normal(0,10,1000)

# 再次用 numpy 裡的函數來計算相關係數
np.corrcoef(x2,y2)
```

```python
# 再看看正相關的 x,y 分布
plt.scatter(x2,y2)
```

### 負相關

```python
# 負相關的散佈圖：https://kknews.cc/news/onkgxaq.html

np.random.seed(1)

# 隨機生成 1000 個介於 0~50 的數 x
x3 = np.random.randint(1,51,1000)

y3 = (x3-49)*(-1) + np.random.normal(0,10,1000)
# (x3-49)*(-1)以後：原本是49的數字，變成0；原本是0的數字，變成49。

np.corrcoef(x3,y3)
```

```python
# 再看看負相關的 x,y 分布
plt.scatter(x3,y3)
```
