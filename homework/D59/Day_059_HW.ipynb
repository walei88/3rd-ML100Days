### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
from sklearn import datasets

digits = datasets.load_digits(n_class=3)
X = digits.data
y = digits.target
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn import decomposition

np.random.seed(5)
%matplotlib inline
```

```python
X.shape
```

```python
pca = decomposition.PCA(n_components=3)

X = pca.fit_transform(X)
```

```python
X.shape
```

```python
fig = plt.figure(1, figsize=(4, 3))
plt.clf() # 筆記：plt.clf()代表清除一個figure
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
plt.cla() # 筆記：plt.clf()代表清除一個axes
# 參考資料：https://blog.csdn.net/tz_zs/article/details/81393098


for name, label in [('0', 0), ('1', 1), ('2', 2)]:
    ax.text3D(X[y == label, 0].mean(),
              X[y == label, 1].mean() + 1.5,
              X[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.7, edgecolor='w', facecolor='w'))
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [0, 1, 2]).astype(np.float)
# np.choose(a, choices)中的a是由int組成的，而且從0開始
ax.scatter(X[:, 0], X[:, 1], X[:, 2] , c=y, cmap=plt.cm.gist_rainbow,
           edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```
