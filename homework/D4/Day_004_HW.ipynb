```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './home-credit-default-risk/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
row_num = format(app_train.shape[0],',') #為讓數字容易閱讀，加上千分位，row_num的type為字符串
clm_num = app_train.shape[1] #clm_num的type為int
print('row數為：%s'%(row_num))
print('column數為：%d'%(clm_num))
```

#### 列出所有欄位

```python
#col_name初始化
col_name = [] 

#==製作col_name的列表，以供後面製作Column Name的字典使用==
for i in range(len(app_train.columns)): 
    col_name.append(app_train.columns[i]) 
    #其實這時候只要下一步print(col_name)就能顯示所有欄位了，
    #不過為了讓顯示的結果更美觀，所以進行以下的處理。

#==製作Column Name的字典，以供後面製作Column Name的表格使用==
dict1 = {
    'Column Name':col_name
}

#==變更顯示的rows的最大值==
pd.set_option("max_rows",len(app_train.columns)) #語法是pd.set_option("max_rows",int)
                                                                      #在這邊的int代表希望顯示的最大rows數。
                                                                      #int必須是正整數。

#==製作Column Name的表格==
df = pd.DataFrame(dict1)
df
```

#### 截取部分資料

```python
#==顯示rows:9~11；columns:'DAYS_EMPLOYED'~'DAYS_ID_PUBLISH' 的資料==
app_train.loc[9:10,'DAYS_EMPLOYED':'DAYS_ID_PUBLISH'] 
```

```python
#==顯示rows:9~11；columns:18~20 的資料==
app_train.iloc[9:11,18:21]
```
